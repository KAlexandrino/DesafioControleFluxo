```python
import json

users = [
    {"id": 1, "name": "Naruto", "news": []},
    {"id": 2, "name": "Hinata", "news": []},
    {"id": 3, "name": "Sasuke", "news": []}
]

print(json.dumps(users, indent=2))
```

```python
import random

def generate_ai_news(user):
    templates = [
        f"{user['name']}, invista hoje para garantir um futuro financeiro seguro!",
        f"Seu dinheiro pode crescer, {user['name']}. Comece a investir agora!",
        f"{user['name']}, pequenos investimentos hoje geram grandes resultados amanhã.",
        f"Invista com sabedoria, {user['name']}, e colha os frutos no futuro!",
        f"{user['name']}, o melhor momento para investir é agora. Não espere!"
    ]
    return random.choice(templates)

for user in users:
    news = generate_ai_news(user)
    print(news)
    user["news"].append({
        "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
        "description": news
    })
```
